## Đọc dữ liệu từ caption được tạo ra

```python
# First line is not header
df_list = []
for cpid in range(4,26,1):

    result_df = pd.read_csv(
        f"./captioning/captioning_MSVD_ResNet152/{cpid}.csv", 
        header=None, names=["VideoID", "caption", "ground_truth"]
    )
    # result_df.info()
    result_df["id"] = result_df["VideoID"].str.split("_").str[0]
    df_list.append(result_df)
    print(f"cpid: {cpid} => # of unique video ids: {len(result_df['id'].unique())}")
```

```python
# Merge all dataframes in a list 
merged_df = pd.concat(df_list, ignore_index=True)
merged_df.info()
```
